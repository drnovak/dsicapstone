```python
import pandas as pd
```

```python
standings_2016 = pd.read_csv('../data/seasons/2016standings.csv')
standings_2015 = pd.read_csv('../data/seasons/2015standings.csv')
standings_2014 = pd.read_csv('../data/seasons/2014standings.csv')
standings_2013 = pd.read_csv('../data/seasons/2013standings.csv')
standings_2012 = pd.read_csv('../data/seasons/2012standings.csv')
```

```python
pd.set_option('display.max_columns',50)
```

```python
standings_2016.head()
```

```python
standings_2015.head()
```

```python
standings_2016 = standings_2016.drop(standings_2016.columns[[0, 2, 3, 7, 11, 12, 13, 14, 15, 18, 24, 25, 26]], axis=1)
standings_2015 = standings_2015.drop(standings_2015.columns[[0, 2, 3, 10, 11, 12, 13, 14, 17]], axis=1)
standings_2014 = standings_2014.drop(standings_2014.columns[[0, 2, 3, 10, 11, 12, 13, 14, 17]], axis=1)
standings_2013 = standings_2013.drop(standings_2013.columns[[0, 2, 3, 10, 11, 12, 13, 14, 17]], axis=1)
standings_2012 = standings_2012.drop(standings_2012.columns[[0, 2, 3, 10, 11, 12, 13, 14, 17]], axis=1)
```

```python
standings_2016['Year'] = 2016
standings_2015['Year'] = 2015
standings_2014['Year'] = 2014
standings_2013['Year'] = 2013
standings_2012['Year'] = 2012
```

```python
standings_2016.head()
```

```python
standings_2015.head()
```

```python
standings = pd.concat([standings_2016, standings_2015, standings_2014, standings_2013, standings_2012], axis=0)
```

```python
standings.dtypes
```

```python
standings.rename(columns={'Tm': 'team', 'W': 'wins', 'L': 'losses', 'W-L%': 'win_loss_pct', 'R': 'runs',
                               'RA': 'runs_allowed', 'Rdiff': 'run_diff','Home': 'home_win_diff', 
                          'Road': 'road_win_diff', '1Run': 'one_run_win_diff', 'vRHP': 'vs_rhp', 'vLHP': 'vs_lhp',
                          '>.500': 'vs_over_500', '<.500': 'vs_under_500', 'Year': 'year'}, inplace=True)
```

```python
standings.head()
```

```python
standings['home_win_diff'] = standings.home_win_diff.apply(lambda x: int(x.split('-')[0]) - int(x.split('-')[1]))
standings['road_win_diff'] = standings.road_win_diff.apply(lambda x: int(x.split('-')[0]) - int(x.split('-')[1]))
standings['one_run_win_diff'] = standings.one_run_win_diff.apply(lambda x: int(x.split('-')[0]) - int(x.split('-')[1]))
standings['vs_rhp'] = standings.vs_rhp.apply(lambda x: int(x.split('-')[0]) - int(x.split('-')[1]))
standings['vs_lhp'] = standings.vs_lhp.apply(lambda x: int(x.split('-')[0]) - int(x.split('-')[1]))
standings['vs_over_500'] = standings.vs_over_500.apply(lambda x: int(x.split('-')[0]) - int(x.split('-')[1]))
standings['vs_under_500'] = standings.vs_under_500.apply(lambda x: int(x.split('-')[0]) - int(x.split('-')[1]))
```

```python
standings.to_csv('standings.csv')
```
