```python
basic_url = 'http://www.baseball-reference.com/leagues/MLB/'

urls = []
for i in range(1980,2017):
    url = basic_url+str(i)+'-standings.shtml'
    urls.append(url)
    
    for url in urls:
        one_year = pd.read_html(url, attrs={"id":'expanded_standings_overall'}, encoding='utf-8')
        one_year[0]['year'] = i
        one_year[0].to_csv('season_{}.csv'.format(i), sep=',', encoding='utf-8')
```
